```python
import openai
import urllib.request
from PIL import Image
import cv2
#from google.colab.patches import cv2_imshow  it will for running the code in colab
```

```python
openai.api_key = "your_api_key_here"
```

```python
img_response = openai.Image.create(
    prompt = 'an astronaut lounging in a tropical resort in space',
    n=1,
    size="512x512")    # 256x256, 512x512, 1024x1024
```

```python
img_response
```

```python
img_response['data'][0]['url']
```

```python
img_url = img_response['data'][0]['url']
img_url
```

```python
urllib.request.urlretrieve(img_url, 'image.png')
```

```python
img = cv2.imread("image.png")

#cv2_imshow(img)
```
